```python
def data (pg):
    table_name = 'Table '+str(pg)
    page_name = str(pg)
    xl = pd.read_excel('college_cutoffs.xlsx',sheet_name = table_name)
    a = xl[list(xl)[0]].dropna()
    b=list(a)
    b[0]=b[-1]
    b=b[:-2]
    st = ''
    for i in range(0,len(b[:])):
        if (b[i] != 'I') and (b[i]!='II'):
            st = st + b[i] +'\n'
    
    ree = re.split(r'[0-9]{9}',st)
    return ree
```
